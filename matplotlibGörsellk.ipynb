```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 2
```

```python
(benimFigure ,benimEksen) = plt.subplots()
benimEksen.plot(numpyDizisi1 , numpyDizisi2,"#b57421", alpha=0.3)
benimEksen.plot(numpyDizisi2, numpyDizisi1, "#1374e8")
```

```python
(yeniFigure , yeniEksen) = plt.subplots()
yeniEksen.plot(numpyDizisi1, numpyDizisi1 +2 , color = "blue", linewidth = 10.0)
yeniEksen.plot(numpyDizisi1, numpyDizisi1 +3 , color="yellow")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 +4 ,color="#ff7cff", linestyle = "-.")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 +5 ,color="#ff7cff", linestyle = ":")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 +6 ,color="#ff7cff", linestyle = "--")

yeniEksen.plot(numpyDizisi1, numpyDizisi1 +7 ,color="#000000", linestyle = "-.", marker="o" ,markersize=9, markerfacecolor="red")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 +8 ,color="#000000", linestyle = "-.", marker="+" ,markersize=8)
```
