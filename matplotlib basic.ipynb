```python
x= [1,2,3]
y=[2,4,6]

plt.plot(x,y)
plt.title("ilk grafik",fontsize=25 ,color="red",loc = "right" ,pad = 50,fontname= "Comic Sans MS")
plt.show()
```

```python
x= [1,2,3]
y=[2,4,6]

plt.plot(x,y)
plt.title("ilk grafik",fontsize=25 ,color="red",loc = "right" ,pad = 50,fontname= "Comic Sans MS")

plt.xlabel(" x ekseni")
plt.ylabel("y ekseni ")
plt.show()
```

```python
x= [1,2,3]
y=[2,4,6]

plt.plot(x,y)
plt.title("ilk grafik",fontsize=25 ,color="red",loc = "right" ,pad = 50,fontname= "Comic Sans MS")

plt.xlabel(" x ekseni", fontsize=25 ,color="red",loc = "right" ,fontname= "Comic Sans MS")
plt.ylabel("y ekseni ", fontsize=25 ,color="red",loc = "bottom" ,fontname= "Comic Sans MS")

plt.show()
```
