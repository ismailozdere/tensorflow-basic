```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
yaslistesi = [10,20,30,40,30,40,50,60,70,75]
```

```python
kiloListesi = [20,60,80,85,86,87,70,90,95,70]
```

```python
numpyYasListesi  = np.array(yaslistesi)
numpyKiloListesi = np.array(kiloListesi)
```

```python
numpyKiloListesi
```

```python
numpyYasListesi
```

```python
plt.plot(numpyYasListesi ,numpyKiloListesi, "g")
plt.xlabel("yas verileri")
plt.ylabel("kilo verileri")
plt.title("kilonun yasa göre değişimi")
#plt.show() # başka  editörlerde kullanılmalı
```

```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 2
```

```python
numpyDizisi2
```

```python
plt.plot(numpyDizisi1,numpyDizisi2, "r*-")
```

```python
plt.subplot(1,2,1)
plt.plot(numpyDizisi1,numpyDizisi2,"r*-")
plt.subplot(1,2,2) # 1sıra 2 sütun olacak 2 sütun çiziliyor 
plt.plot(numpyDizisi2,numpyDizisi1,"y--")
```

```python
benimFigure = plt.figure()
```

```python
benimFigure = plt.figure()
figureAxes = benimFigure.add_axes([0.2,0.2,0.9,0.9])
figureAxes.plot(numpyDizisi1,numpyDizisi2,"r")
figureAxes.set_xlabel("x ekseni veri ismi ")
figureAxes.set_ylabel("y ekseni veri ismi")
figureAxes.set_title("grafik başlığı")
```

```python
figure2  = plt.figure()

eksen1 = figure2.add_axes([0.1,0.1,0.9,0.9])
eksen2 = figure2.add_axes([0.2,0.4,0.3,0.3])

eksen1.plot(numpyDizisi1,numpyDizisi2, "r")
eksen1.set_xlabel("1. x ekseni")
eksen1.set_ylabel("1. y ekseni")
eksen1.set_title("ana grafik başlık")

eksen2.plot(numpyDizisi2, numpyDizisi1, "g")
eksen2.set_xlabel("2. x ekseni")
eksen2.set_ylabel("2. y ekseni")
eksen2.set_title("küçük grafik başlık")
```

```python
(benimFigure , benimEksenler) = plt.subplots(nrows = 1 , ncols = 2)
type(benimEksenler)

for eksen in benimEksenler:
    eksen.plot(numpyDizisi1 , numpyDizisi2 ,"r")
    eksen.set_xlabel("x ekseni")
     
plt.tight_layout()
```

```python
yeniFigure = plt.figure( dpi= 200)
#figsize=(9,9)

yeniEksen  = yeniFigure.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizisi1,numpyDizisi2, label = "numpyDizisi **2")
yeniEksen.plot(numpyDizisi1 , numpyDizisi1 ** 3 , label ="numpyDizisi ** 3")
yeniEksen.legend(loc = 2)
```

```python
yeniFigure.savefig("benimfigur.png",dpi=250)
```
